# Familiarising yourself with the Enviornment

## 1. Logon to the enviornment

Login to the Lab environment using this link [Lab Environment](https://login.cdpworkshops.cloudera.com/auth/realms/field-marketing-emea-eu/protocol/saml/clients/cdp-sso).

`https://login.cdpworkshops.cloudera.com/auth/realms/field-marketing-emea-eu/protocol/saml/clients/cdp-sso`

![image10.jpeg](assets/image18.jpeg)

Enter your assigned username and the password provided. For example if you are assigned **user003** enter **user003** as the user name. Click the **Sign In** button. This will open the Cloudera Control Plane. 

## 2. Navigate to Cloudera Workbench

In the Cloudera Control Plane, click on the **Cloudera AI** tile.

![image10.jpeg](assets/image19.jpeg)

Select **AI Workbenches** in the left menu, and select the HOL workbench.

![image10.jpeg](assets/image20.jpeg)

## 3. Open your assigned user Project 

Click on the Project tile that relates to your assigned username. For example if you are assigned **user006** click on the **user006_hol** tile.

![image10.jpeg](assets/image21.jpeg)

## 4. Start a JuypterLab Session 

Within your assigned Project, click on **Sessions** in the left menu and then **New Session** OR click **New Session** at the top right of the UI.

![image10.jpeg](assets/image22.jpeg)

Accept all the defaults and click the **Start button**.

![image10.jpeg](assets/image23.jpeg)

You will see the following screen were your session is initialising.

![image10.jpeg](assets/image24.jpeg)

Once you session has initialised, a **Code Connection Snippet** popup will display. Click the **Close** button.

![image10.jpeg](assets/image25.jpeg)

## 5. Working with JuypterLab Notebooks 

You will now be in a JupyterLab Notebook. Double-click on the lab Notebooks in the left menu to open them. Click on the different code cells to focus on them (when selected the cell hightlights with a blue surround and blue line on the left). At the top of notebooks, where requested, make sure to update the username to your assigned username. For example if you are assigned **user003** change the code to **username = "user003"**. 

![image10.jpeg](assets/image26.jpeg)

To execute code block, again make sure you have focus on the code cell block by clicking on it. To execute the code click on the **run** button in the top menu. As the cell executes you will observe the output below the cell.

![image10.jpeg](assets/image27.jpeg)
